```python
1.2 Database: SQL Query II
==========================
# 1. Menulis Queri SQL untuk menampilkan Total Bencana Alam dan Kerusakan Ekonomi untuk setiap jenis Bencana (Entity)
# Queri ini menggunakan JOIN untuk menggabungkan kedua tabel berdasarkan kolom "Entity" dan "Year" yang ada di kedua tabel (2 tabel berbeda)

SELECT nd."Entity", SUM(nd."Number of reported natural disasters") AS "Total Natural Disasters", SUM(ed."Total economic damage from natural disasters") AS "Total Economic Damage"
FROM "datasetjumlahbencanaalam" AS nd
JOIN "datasettotalekonomi" AS ed ON nd."Entity" = ed."Entity" AND nd."Year" = ed."Year"
GROUP BY nd."Entity"
ORDER BY nd."Entity";
```

2. Merangkum Data per 10 Tahun, Menghitung Averages dan Sums for Natural Disaster & Economic Damage

Query ini untuk merangkum data per 10 tahun dan menghitung rata-rata dan jumlah bencana alam dan kerusakan ekonomi

```python
SELECT
  (nd."Year" / 10) * 10 AS "Tahun",
  AVG(nd."Number of reported natural disasters") AS "Average Natural Disasters",
  SUM(nd."Number of reported natural disasters") AS "Total Natural Disasters",
  AVG(ed."Total economic damage from natural disasters") AS "Average Economic Damage",
  SUM(ed."Total economic damage from natural disasters") AS "Total Economic Damage"
FROM "datasetjumlahbencanaalam" AS nd
JOIN "datasettotalekonomi" AS ed ON nd."Entity" = ed."Entity" AND nd."Year" = ed."Year"
GROUP BY "Tahun"
ORDER BY "Tahun";
```

dengan menggabungkan kedua data sehingga menjadi satu kesatuan data

```python
import pandas as pd

# Load dataset 1
url1 = "https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/number-of-natural-disaster-events.csv"
data1 = pd.read_csv(url1)

# Load dataset 2
url2 = "https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/economic-damage-from-natural-disasters.csv"
data2 = pd.read_csv(url2)

merged_data = pd.merge(data1, data2, on=["Entity", "Year"])
print(merged_data.head())
```

MENGHITUNG MEAN, MEDIAN DAN STANDARD DEVIASI

```python
import pandas as pd
# Menghitung mean (rata-rata)
mean_natural_disasters = merged_data["Number of reported natural disasters (reported disasters)"].mean()
mean_economic_damage = merged_data["Total economic damage from natural disasters (US$)"].mean()

# Hitung median (nilai tengah) untuk kolom yang diperlukan
median_natural_disasters = merged_data["Number of reported natural disasters (reported disasters)"].median()
median_economic_damage = merged_data["Total economic damage from natural disasters (US$)"].median()

# Menghitung standar deviasi
std_deviation_natural_disasters = merged_data["Number of reported natural disasters (reported disasters)"].std()
std_deviation_economic_damage = merged_data["Total economic damage from natural disasters (US$)"].std()

print("JADI NILAI DARI MEAN, MEDIAN, DAN STANDAR DEVIASI TIAP DATASET ADALAH:\n")

print(f"1. Mean (rata-rata) dari Data Jumlah Bencana Alam: {mean_natural_disasters}")
print(f"2. Mean (rata-rata) dari Data Total Kerusakan Ekonomi (US$): {mean_economic_damage}")
print(f"3. Median (nilai tengah) dari Data Jumlah Bencana Alam: {median_natural_disasters}")
print(f"4. Median (nilai tengah) dari Data Total Kerusakan Ekonomi (US$): {median_economic_damage}")
print(f"5. Standard Deviation (standar deviasi) dari Data Jumlah Bencana Alam: {std_deviation_natural_disasters}")
print(f"6. Standard Deviation (standar deviasi) dari Data Total Kerusakan Ekonomi (US$): {std_deviation_economic_damage}")
```

Visualisasi Hasil Hitung Mean, Median, dan Standar Deviasi Tiap Variabel

```python
import matplotlib.pyplot as plt

data_labels = ["Mean Natural Disasters", "Median Natural Disasters", "Std. Deviation Natural Disasters",
               "Mean Economic Damage (US$)", "Median Economic Damage (US$)", "Std. Deviation Economic Damage (US$)"]

data_values = [mean_natural_disasters, median_natural_disasters, std_deviation_natural_disasters,
               mean_economic_damage, median_economic_damage, std_deviation_economic_damage]

# Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(data_labels, data_values, color=['blue', 'green', 'red', 'blue', 'green', 'red'])
plt.xlabel('Statistics')
plt.ylabel('Values')
plt.title('Mean, Median, and Standard Deviation')
plt.xticks(rotation=45, ha="right")  # Mengatur rotasi dan posisi label
plt.tight_layout()  # Mengatur jarak antar label agar sejajar


# Menambahkan nilai di atas setiap batang
for bar, value in zip(bars, data_values):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, value, str(round(value, 2)), fontsize=10, color='black', ha='center')
plt.show()
```

MEMBUAT K-MEANS CLUSTERING

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Pilih fitur-fitur yang akan digunakan untuk clustering
data_for_clustering = merged_data[['Number of reported natural disasters (reported disasters)', 'Total economic damage from natural disasters (US$)']]

# Standarisasi data jika perlu
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_for_clustering_scaled = scaler.fit_transform(data_for_clustering)

# Menentukan jumlah cluster (kelompok) dengan Elbow Method
wcss = []  # Within-Cluster-Sum-of-Squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data_for_clustering_scaled)
    wcss.append(kmeans.inertia_)

# Plot Elbow Method untuk menentukan jumlah cluster yang tepat
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal Cluster Number')
plt.grid(True)
plt.show()

# Berdasarkan hasil Elbow Method, pilih jumlah cluster yang optimal dan lakukan clustering
optimal_num_clusters = 3  # Gantilah sesuai dengan hasil Elbow Method Anda
kmeans = KMeans(n_clusters=optimal_num_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(data_for_clustering_scaled)

# Tambahkan label kluster ke data
merged_data['Cluster'] = cluster_labels

# Visualisasikan hasil clustering
plt.figure(figsize=(8, 6))
plt.scatter(data_for_clustering_scaled[:, 0], data_for_clustering_scaled[:, 1], c=cluster_labels, cmap='rainbow')
plt.xlabel('Number of Reported Natural Disasters (Standardized)')
plt.ylabel('Total Economic Damage (US$) (Standardized)')
plt.title('K-means Clustering Result')
plt.show()
```

Analisis Regresi untuk memprediksi Kerusakan Ekonomi berdasarkan Jumlah Bencana Alam

```python
import statsmodels.api as sm
X = merged_data['Number of reported natural disasters (reported disasters)']
y = merged_data['Total economic damage from natural disasters (US$)']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
print(model.summary())
```

Penjelasan Hasil:
Hasil analisis regresi menunjukkan sejauh mana model yang digunakan dalam analisis regresi dapat menjelaskan atau memprediksi kerusakan ekonomi berdasarkan jumlah bencana alam.
Nilai R-squared sebesar 0.488 mengindikasikan bahwa sekitar 48.8% variabilitas kerusakan ekonomi dapat dijelaskan oleh jumlah bencana alam dalam model ini.
Dengan kata lain, model ini memberikan gambaran tentang hubungan positif antara jumlah bencana alam dan kerusakan ekonomi, meskipun masalah asumsi perlu diperhatikan.

Regresi Linear untuk memprediksi kerusakan ekonomi berdasarkan jumlah bencana alam dengan menggunakan regresi linear

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Pisahkan data berdasarkan variabel jumlah bencana alam dan kerusakan ekonomi
X = merged_data[['Number of reported natural disasters (reported disasters)']]
y = merged_data['Total economic damage from natural disasters (US$)']

# Membuat model regresi linear
model = LinearRegression()
model.fit(X, y)

# Prediksi kerusakan ekonomi berdasarkan jumlah bencana alam
y_pred = model.predict(X)

# Visualisasi hasil regresi
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Number of Reported Natural Disasters')
plt.ylabel('Total Economic Damage (US$)')
plt.legend()
plt.title('Linear Regression: Natural Disasters vs. Economic Damage')
plt.show()

# Print koefisien regresi dan intercept
print("Koefisien Regresi (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```


Hasil Koefisien Regresi (Slope) yang positif sebesar 285,679,553.52 menunjukkan bahwa setiap peningkatan satu unit dalam jumlah bencana alam yang dilaporkan (misalnya, dari satu bencana tambahan) berkontribusi pada peningkatan sekitar $285,679,553.52

dalam kerusakan ekonomi. Intercept yang negatif (-$1,563,131,830.98) adalah perkiraan kerusakan ekonomi ketika jumlah bencana alam yang dilaporkan adalah nol. Artinya, bahkan ketika tidak ada bencana alam yang dilaporkan, masih ada kerusakan ekonomi awal yang ditunjukkan oleh nilai intercept tersebut.

Visualisasi untuk hasil korelasi antara jumlah bencana alam dan kerusakan ekonomi menggunakan scatter plot. Berisi titik-titik data dan garis regresi linear untuk memvisualisasikan hubungan antara dua variabel

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Data
x = merged_data['Number of reported natural disasters (reported disasters)']
y = merged_data['Total economic damage from natural disasters (US$)']

# Hitung korelasi Pearson
correlation, p_value = stats.pearsonr(x, y)

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.7)
plt.xlabel('Number of Reported Natural Disasters')
plt.ylabel('Total Economic Damage (US$)')
plt.title('Scatter Plot: Natural Disasters vs. Economic Damage')
plt.grid(True)

# Garis regresi linear
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
plt.plot(x, slope * x + intercept, color='red', label=f'Regression Line (Correlation = {correlation:.2f})')
plt.legend()
plt.show()
```
